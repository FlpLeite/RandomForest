# Eda
Este notebook realiza a etapa de **eda** no pipeline de Machine Learning para o dataset Wine Quality.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../datasets/wine_processed.csv")
df.head()
```

## Importação e leitura do dataset processado

```python
df.hist(bins=20, figsize=(15,10))
plt.show()
```

## Histogramas das variáveis

```python
corr = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

## Matriz de correlação

```python
sns.countplot(data=df, x="quality")
plt.title("Distribuição das notas de qualidade")
plt.show()
```

## Distribuição da qualidade

```python
for col in df.columns[:-1]:  # exceto 'quality'
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col], showfliers=False)
    plt.title(f"Boxplot - {col}")
    plt.show()
```
